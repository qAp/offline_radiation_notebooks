#### Load RRTMG (python interface) results 

```python
pathout_rrtmg = '/nuwa_cluster/home/jackyu/radiation/rrtmg/LW/rrtmg_lw_v4.85/column_model/runs/mls70_O3/OUTPUT_RRTM'
outrad = sum_OUTPUT_RRTM_over_wave_numbers(readfrom = pathout_rrtmg, V1 = 0, V2 = 3000)
pres_rrtmg_py = .5 * (outrad.values[1:, 0] + outrad.values[: -1, 0])
cor_rrtmg_py = outrad.ix[outrad.index[1:], 'cooling_rate']
```

#### Plot cooling rates

```python
matplotlib.rcParams.update({'font.size': 20.})
fig = plt.figure(figsize = (13, 13))
ax = fig.add_subplot(111,
                     xlabel = 'cooling rate [deg/day]', xlim = [-.5, 2.5],
                     ylabel = 'pressure [mb]', ylim = [1e-4, 2e3],
                     title = 'Cooling rates. $O_3$')

lines = ax.semilogy(cor_lbl, pres_lbl,\
                    cor_clirad, pres_clirad,\
                    cor_lblrtm, pres_lblrtm,\
                    cor_rrtmg, pres_rrtmg,
                    cor_rrtmg_py, pres_rrtmg_py)

colours = ['b'] * 2 + 2 * ['g'] + ['r']
linestyles = ['-', '--'] * 2 + ['--']
[plt.setp(line, linestyle = style, color = colour, linewidth = 1.5)\
for line, style, colour in zip(lines, linestyles, colours)]

plt.legend(['CRD', 'CLIRAD', 'AER LBLRTM', 'AER RRTMG (1)', 'AER RRTMG (2)'])
plt.grid(b = True)
plt.gca().invert_yaxis()

path_fig = 'cor_mls_O3.pdf'
plt.savefig(path_fig)
```

# Cooling rate difference above ~ 100 mb in RRTMG's 16 bands: AER LBLRTM vs AER RRTMG

```python
orad_rrtmg = rtmtools.rrtmg.lw.wrangle.OUTPUT_RRTM_to_pandasPanel(readfrom = pathout_rrtmg, signed_fluxes=True, cooling_rate=True)
```

```python
for V1, V2 in orad_rrtmg.items:
    print('Wave number band {} ~ {} cm-1'.format(V1, V2))
    df_rrtmg = sum_OUTPUT_RRTM_over_wave_numbers(readfrom = pathout_rrtmg, V1 = V1, V2 = V2)
    df_lblrtm = aerutils.sum_OUTPUT_RADSUM_over_wave_numbers(readfrom = pathout_lblrtm, V1 = V1, V2 = V2)
    print(df_rrtmg[df_rrtmg['pressure'] > 100]['cooling_rate'] - df_lblrtm[df_lblrtm['pressure'] > 100]['cooling_rate'])
    print()
    print()
```
