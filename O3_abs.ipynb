```python
%matplotlib inline
import os
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
```

```python
def wave_length2number(wavelength):
    '''
    wavelength in metres
    wavenumber in cm-1
    '''
    return 1e-2 / wavelength
```

```python
# O3 absorption data used by CRD-SW

path_crd = '/nuwa_cluster/home/jackyu/radiation/crd/SW/src/o3abs.data'
o3abs_crd = pd.read_csv(path_crd, header = None, sep = r'\s+', names = ['wavenumber', 'absorption', 'rayleigh'])
```

```python
matplotlib.rcParams.update({'font.size': 16})
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111,
                     xlabel = 'wavenumber [$cm^{-1}$]')
ax.plot(o3abs_crd['wavenumber'], o3abs_crd['absorption'])
ax.set_yscale('log')

ax.xaxis.get_major_formatter().set_powerlimits((0, 1))
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111,
                     xlabel = 'wavenumber [$cm^{-1}$]')
ax.plot(o3abs_crd['wavenumber'], o3abs_crd['rayleigh'])

ax.xaxis.get_major_formatter().set_powerlimits((0, 1))
```
