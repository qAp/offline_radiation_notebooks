# In RRTMG's wavenubmer bands

```python
levels =  [70, 30, 0]
attrs = ['pressure', 'flux_up', 'flux_down', 'net_flux', 'heating_rate']
name_molecule = '$O_3$'
```

```python
crd_wrt_rrtmg = aeranalyse.lines2bands(pnl_crd, wbands = pnl_rrtmg.items)
crd_wrt_rrtmg_lvls = crd_wrt_rrtmg.ix[:, levels, :]
```

```python
rrtmg_wrt_rrtmg = pnl_rrtmg.ix[:, :, attrs]
rrtmg_wrt_rrtmg = aeranalyse.normalise_by_TOA_flux_down(rrtmg_wrt_rrtmg, 
                                                        normalise_to = crd_wrt_rrtmg)
rrtmg_wrt_rrtmg_lvls = rrtmg_wrt_rrtmg.ix[:, levels, :]
```

```python
df1 = aeranalyse.sum_OUTPUT_RADSUM_over_wbands(crd_wrt_rrtmg, 
                                               V1 = crd_wrt_rrtmg.items[0][0],
                                               V2 = crd_wrt_rrtmg.items[-1][-1])
df2 = aeranalyse.sum_OUTPUT_RADSUM_over_wbands(rrtmg_wrt_rrtmg, 
                                               V1 = rrtmg_wrt_rrtmg.items[0][0],
                                               V2 = rrtmg_wrt_rrtmg.items[-1][-1])

df1_lvls = df1.ix[levels, :]
df2_lvls = df2.ix[levels, :]

viz.tabulate_difference(dfs = [df1_lvls, df2_lvls], 
                        names = ['CRD', 'RRTMG'],
                        title = 'Wavenumber range: {} ~ {} cm-1'.\
                        format(pnl_rrtmg.items[0][0], pnl_rrtmg.items[-1][-1]))

viz.plot_pres_vs_flux_down(dfs = [df1, df2],
                           names = ['CRD', 'RRTMG'], 
                           linestyles = ['-', '--'], 
                           colours = ['b', 'g'],
                           pres_scale = 'log',
                           title = 'Flux down {}. {} ~ {} cm-1'.\
                           format(name_molecule, pnl_rrtmg.items[0][0], pnl_rrtmg.items[-1][-1]))

display.display(plt.gcf()); plt.close()

viz.plot_pres_vs_hrcr(dfs = [df1, df2],
                      names = ['CRD', 'RRTMG'],
                      linestyles = ['-', '--'],
                      colours = ['b', 'g'],
                      pres_scale = 'log',
                      title = 'Heating rate {}. {} ~ {} cm-1'.\
                      format(name_molecule, pnl_rrtmg.items[0][0], pnl_rrtmg.items[-1][-1]))

display.display(plt.gcf()); plt.close()

print()
print('--------------------------------------------------------------------')
print()


for item in pnl_rrtmg.items:
    df1_lvls = crd_wrt_rrtmg_lvls[item]
    df2_lvls = rrtmg_wrt_rrtmg_lvls[item]
   
    df1 = crd_wrt_rrtmg[item]
    df2 = rrtmg_wrt_rrtmg[item]
    
    viz.tabulate_difference(dfs = [df1_lvls, df2_lvls], 
                            names = ['CRD', 'RRTMG'],
                            title = 'Wavenumber band: {} ~ {} cm-1'.format(*item))
    
    
    
    viz.plot_pres_vs_flux_down(dfs = [df1, df2],
                               names = ['CRD', 'RRTMG'],
                               linestyles = ['-', '--'], 
                               colours = ['b', 'g'],
                               pres_scale = 'log',
                               title = 'Flux down {}. {} ~ {} cm-1'.format(name_molecule, *item))
    
    display.display(plt.gcf()); plt.close()

    viz.plot_pres_vs_hrcr(dfs = [df1, df2],
                          names = ['CRD', 'RRTMG'],
                          linestyles = ['-', '--'], 
                           colours = ['b', 'g'],
                           pres_scale = 'log',
                          title = 'Heating rate {}. {} ~ {} cm-1'.format(name_molecule, *item))
    
    display.display(plt.gcf()); plt.close()
    
    print()
    print('--------------------------------------------------------------------')
    print()
```

# In CLIRAD's wavenumber bands

```python
cliradsw_wbands = rtmtools.clirad.sw.info.wavenumber_bands()
```

```python
levels = [70, 30, 0]
attrs = ['pressure', 'flux_up', 'flux_down', 'net_flux', 'heating_rate']
name_molecule = '$O_3$'
wbands = list(range(1, 8 + 1))
wranges = {wband: cliradsw_wbands[wband] for wband in wbands}
```

```python
crd_wrt_clirad = rtmtools.clirad.sw.analyse.lines2bands(pnl_crd, wbands = wranges)
crd_wrt_clirad_lvls = crd_wrt_clirad.ix[:, levels, :]
```

```python
clirad_wrt_clirad = pnl_clirad.ix[:, :, attrs]
aeranalyse.normalise_by_TOA_flux_down(clirad_wrt_clirad,
                                      normalise_to = crd_wrt_clirad)
clirad_wrt_clirad_lvls = clirad_wrt_clirad.ix[:, levels, :]
```

```python
df1 = cliradanalyse.sum_OUTPUT_CLIRAD_over_wbands(crd_wrt_clirad, wbands = range(1, 8 + 1))
df2 = cliradanalyse.sum_OUTPUT_CLIRAD_over_wbands(clirad_wrt_clirad, wbands = range(1, 8 + 1))

df1_lvls = df1.ix[levels, :]
df2_lvls = df2.ix[levels, :]

wlabel = ' '.join(['{} ~ {}'.format(*interval) \
                   for key, intervals in wranges.items() \
                   for interval in intervals]) + ' cm-1'

viz.tabulate_difference(dfs = [df1_lvls, df2_lvls], 
                        names = ['CRD', 'CLIRAD'],
                        title = 'Wavenumber range: bands {} ~ {}, {}'.\
                        format(wbands[0], wbands[-1], wlabel))
    
viz.plot_pres_vs_flux_down(dfs = [df1, df2],
                           names = ['CRD', 'CLIRAD'],
                           linestyles = ['-', '--'],
                           colours = ['b', 'b'],
                           pres_scale = 'linear',
                           title = 'Flux down {}'.format(name_molecule))

display.display(plt.gcf()); plt.close()

viz.plot_pres_vs_hrcr(dfs = [df1, df2],
                      names = ['CRD', 'CLIRAD'],
                      linestyles = ['-', '--'],
                      colours = ['b', 'b'],
                      pres_scale = 'log',
                      title = 'Heating rate {}'.format(name_molecule))

display.display(plt.gcf()); plt.close()

print()
print('--------------------------------------------------------------------')
print()


for wband in wbands:
    df1_lvls = crd_wrt_clirad_lvls[wband]
    df2_lvls = clirad_wrt_clirad_lvls[wband]
    
    df1 = crd_wrt_clirad[wband]
    df2 = clirad_wrt_clirad[wband]
    
    wlabel = ' '.join(['{} ~ {}'.format(*interval) for interval in wranges[wband]]) + ' cm-1'
    
    viz.tabulate_difference(dfs = [df1_lvls, df2_lvls], 
                            names = ['CRD', 'CLIRAD'],
                            title = 'Wavenumber band: {}, {}'.format(wband, wlabel))
    
    viz.plot_pres_vs_flux_down(dfs = [df1, df2],
                               names = ['CRD', 'CLIRAD'],
                               linestyles = ['-', '--'],
                               colours = ['b', 'b'],
                               pres_scale = 'linear',
                               title = 'Flux down {} {}'.format(name_molecule, wlabel))
    
    display.display(plt.gcf()); plt.close()

    viz.plot_pres_vs_hrcr(dfs = [df1, df2],
                          names = ['CRD', 'CLIRAD'],
                          linestyles = ['-', '--'],
                          colours = ['b', 'b'],
                          pres_scale = 'log',
                          title = 'Heating rate {} {}'.format(name_molecule, wlabel))
    
    display.display(plt.gcf()); plt.close()
    
    print()
    print('--------------------------------------------------------------------')
    print()
```
