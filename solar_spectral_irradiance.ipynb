```python
%matplotlib inline
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
def wave_length2number(wavelength):
    '''
    wavelength in metres
    wavenumber in cm-1
    '''
    return 1e-2 / wavelength
```

```python
# SORCE
path_sorce = '/nuwa_cluster/home/jackyu/radiation/sorce_sim_L3_c24h_0240nm_2413nm_20030414_20130315.txt'
```

```python
sorce_data = pd.read_csv(path_sorce, skiprows = 79, header = None, 
                         index_col = list(range(4)), sep = r'\s+',
                         names = ['instrument_mode_id',
                                  'data_version',
                                  'irradiance',
                                  'irradiance_uncertainty',
                                  'quality'])
```

```python
sorce_data.index.names = ['nominal_date_yyyymmdd',
                          'nominal_date_jdn',
                          'min_wavelength',
                          'max_wavelength']
```

```python
sorce_data = sorce_data.reset_index()
```

```python
# Look at the last day
yyyymmdd, jdn = 20130315.5, 2456367
sorce_lastday = sorce_data.loc[\
                               (sorce_data['nominal_date_yyyymmdd'] == yyyymmdd) &
                               (sorce_data['nominal_date_jdn'] == jdn)].loc[:, ['min_wavelength', 'irradiance']]

sorce_lastday['wavenumber'] = wave_length2number(sorce_lastday['min_wavelength'] * 1e-9)

sorce_lastday = sorce_lastday.sort('wavenumber')
sorce_lastday = sorce_lastday.reset_index()
sorce_lastday = sorce_lastday[['min_wavelength', 'wavenumber', 'irradiance']]

# Convert W m^{-2} (nm)^{-1} to W m^{-2} cm
sorce_lastday['irradiance_v'] = sorce_lastday['min_wavelength']**2 / 1e7 * sorce_lastday['irradiance']
```

```python
# CRD's 'new.dat'
path_lee = '/nuwa_cluster/home/jackyu/radiation/crd/SW/src/new.dat'
crd_data = pd.read_csv(path_lee, sep = r'\s+',  
                       header = None,
                       names = ['wavenumber', 'irradiance'])
```
