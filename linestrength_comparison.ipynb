#Comparison of integrated line strengths

In this section, a comparison is made of the integrated line strength and number of lines in each wavenumber band, between AER LBLRTM, CRD and the original HITRAN 2012 files.

```python
interests = ['num_of_lines', 'integrated_intensity']
names_molecules = ['H2O', 'CO2', 'O3']
names_sources = ['AER LBLRTM', 'CRD', 'HITRAN 2012']
```

```python
#Compare H2O in table
compare_H2O = pd.concat([LS_AER_H2O.loc[:, interests], \
                        LS_CRD_H2O.loc[:, interests], 
                        LS_H_H2O.loc[:, interests]], 
                          axis = 1, 
                          keys = names_sources)
# Compare CO2 in table
compare_CO2 = pd.concat([LS_AER_CO2.loc[:, interests], \
                        LS_CRD_CO2.loc[:, interests], 
                        LS_H_CO2.loc[:, interests]], 
                          axis = 1, 
                          keys = names_sources)
# Compare O3 in table
compare_O3 = pd.concat([LS_AER_O3.loc[:, interests], \
                        LS_CRD_O3.loc[:, interests], 
                        LS_H_O3.loc[:, interests]], 
                          axis = 1, 
                          keys = names_sources)
```

```python
compare_CO2.sum()
```

```python
compares = [compare_H2O, compare_CO2, compare_O3]
with pd.ExcelWriter('integrated_line_intensity_comparisons.xlsx') as writer:
    [df.to_excel(writer, sheet_name = name) for df, name in zip(compares, names_molecules)]
```

```python
# plot H2O comparison (AER vs HITRAN 2012)
names_sources = ['AER LBLRTM', 'HITRAN 2012']
matplotlib.rcParams.update({'font.size': 20.})
fig = plt.figure(figsize = (12, 12))
ax = fig.add_subplot(111,
                     title = 'H2O',
                     ylabel = 'integrated line strength',
                     xlabel = 'starting wavenumber of band [$cm^{-1}$]')

plotargs = list(\
                itertools.chain(\
                                *((compare_H2O.index,\
                                   compare_H2O[(name, 'integrated_intensity')])\
                      for name in names_sources)\
                                )\
)

lines = ax.plot(*plotargs)

colours = ['g', 'r']
linestyles  = ['--', '--']
[plt.setp(line, color = colour, linestyle = linestyle, linewidth = 1.5) \
 for line, colour, linestyle in zip(lines, colours, linestyles)]

ax.xaxis.get_major_formatter().set_powerlimits([0, 3])
ax.set_yscale('log')
ax.set_xlim(left = 0.)
plt.legend(names_sources, loc = 'best')
plt.grid(b = True)

#plt.savefig('LS_H2O.pdf')
```
